```python
import pandas as pd

# CSV dosyanı okutuyorsun
df = pd.read_csv("sahibindenx.csv")

# Sadece işimize yarayan sütunları seçip yeni bir DataFrame oluşturuyoruz
df_temiz = df[["searchResultsAttributeValue", "searchResultsAttributeValue 2", 
               "classified-price-container", "searchResultsLocationValue"]].copy()

# Sütun isimlerini makul, anlaşılır isimlerle değiştiriyoruz
df_temiz.columns = ["Metrekare", "Oda", "Fiyat", "Konum"]

# Fiyat sütunundaki " TL" yazısını ve binlik ayracı olan "." işaretini siliyoruz
df_temiz["Fiyat"] = df_temiz["Fiyat"].str.replace(" TL", "").str.replace(".", "")

# Fiyatı matematiksel işlem yapılabilmesi için Integer (tam sayı) tipine dönüştürüyoruz
df_temiz["Fiyat"] = df_temiz["Fiyat"].astype(int)

# Temizlenmiş verimizin ilk birkaç satırına bakıyoruz
print(df_temiz.head())
```

```python
df.isnull().sum()
```

```python
# Oda sütunundaki "3+1" gibi değerleri toplar (Örn: 4 yapar)
df_temiz['Oda'] = df_temiz['Oda'].apply(lambda x: sum(int(i) for i in str(x).split('+')))

# Mahalle isimlerini ayırma işlemleri için de df_temiz kullanmalısın
df_temiz['Konum'] = df_temiz['Konum'].str.replace("Ataevler", "Ataevler ")
df_temiz['Konum'] = df_temiz['Konum'].str.replace("Beşevler", "Beşevler ")
```

```python
print(df_temiz['Oda'].unique())
```

```python
# Oda sayısını hesaplayan akıllı bir fonksiyon yazıyoruz
def oda_hesapla(deger):
    deger = str(deger).lower().strip() # Yazıyı küçült ve boşlukları sil
    
    if deger == 'nan':
        return None # Boş verileri None yapıyoruz
    elif 'stüdyo' in deger:
        return 1 # Stüdyo daireler 1 oda sayılır
        
    try:
        # "3+1" formatında olanları topla
        return sum(int(i) for i in deger.split('+'))
    except ValueError:
        # Eğer "10 ve üzeri" gibi hesaplanamayan bir şey gelirse None (boş) yap
        return None

# Yazdığımız bu fonksiyonu Oda sütununa uyguluyoruz
df_temiz['Oda'] = df_temiz['Oda'].apply(oda_hesapla)

# Eğer hesaplanamayan (None olan) ilanlar oluştuysa onları veri setinden siliyoruz
df_temiz = df_temiz.dropna(subset=['Oda'])

# EKLENEN KISIM: Boşluklar silindikten sonra sütunu tam sayı (integer) formatına zorluyoruz
df_temiz['Oda'] = df_temiz['Oda'].astype(int)
```

```python
print(df_temiz[['Oda', 'Konum']].head(10))
```

```python
# Belde sütununa veriyi çekiyoruz (Bu kısım aynı)
df_temiz['Belde'] = df_temiz['Konum'].str.extract('(Ataevler|Beşevler)')

# Sadece cümlenin en başındaki (^) Ataevler veya Beşevler kelimelerini sileriz
df_temiz['Konum'] = df_temiz['Konum'].str.replace('^(Ataevler|Beşevler)', '', regex=True).str.strip()
```

```python
print(df_temiz[['Belde', 'Konum']].head(10))
```

```python
print(df_temiz.isnull().sum())
```

```python
# Eğer hücrede sadece boşluk karakteri varsa (veya tamamen boşsa) bunu Pandas'ın anlayacağı 'None' (boş değer) formatına çeviriyoruz
df_temiz['Konum'] = df_temiz['Konum'].replace(r'^\s*$', None, regex=True)

# 'Konum' sütununda boş değer (None/NaN) olan tüm satırları veri setinden siliyoruz
df_temiz = df_temiz.dropna(subset=['Konum'])
```

```python
df_temiz.to_csv('final.csv', index=False)
```
